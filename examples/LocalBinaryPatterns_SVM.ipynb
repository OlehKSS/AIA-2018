```python
import cv2
import matplotlib.pyplot as plt
from immas import MammogramImage , preprocessing , segmentation
from immas.io import read_dataset
from immas.basic_functions import show_image_plt

from immas import preprocessing

from skimage.feature import local_binary_pattern
import numpy as np
```

```python
def testPreProcessing (img):
    img = preprocessing.morphoEnhancement(img,2,10.0)
    img = preprocessing.waveletTransform(img)
    return img
```

```python
from PIL import Image
import glob
# Extracting Local Binary Patterns
def extractLBP (image):
    lbp = local_binary_pattern(image, n_points, radius, METHOD)
    n_bins = int(lbp.max() + 1)
    hist, _ = np.histogram(lbp, normed=True, bins=n_bins, range=(0, n_bins))
    
    return hist
```

```python
# Create training set
labels = []
tr_data = []

# Parameters for LBP
radius = 8
n_points = 24
METHOD = 'uniform'

for filename in glob.glob('../dataset/regions_of_interest/MASS/training/*.tif'):
    im = cv2.imread(filename, 0)
    # Preprocess image
    im = testPreProcessing(im)
    # Describe with LBP
    im_hist = extractLBP(im)
    # add to the training set
    tr_data.append(im_hist)
    labels.append('P')
    print(filename,'P')
    
for filename in glob.glob('../dataset/regions_of_interest/NO_MASS/training/*.tif'):
    im = cv2.imread(filename, 0)
    # Preprocess image
    im = testPreProcessing(im)
    # Describe with LBP
    im_hist = extractLBP(im)
    # add to the training set
    tr_data.append(im_hist)
    labels.append('N')
    print(filename,'N')
    
     
```

```python
from sklearn.svm import LinearSVC

# create Suport Vector Machine model
model = LinearSVC(C=100.0, random_state=42)
model.fit(tr_data, labels)
```

```python
# loop over testing images with masses
for filename in glob.glob('../dataset/regions_of_interest/MASS/test/*.tif'): 
    im = cv2.imread(filename, 0)
    # Preprocess image
    im = testPreProcessing(im)
    # describe
    hist = extractLBP(im)
    # classify
    prediction = model.predict(hist.reshape(1,-1))
    print("Prediction: ",prediction)
    
```

```python
# loop over testing images w/o masses
for filename in glob.glob('../dataset/regions_of_interest/NO_MASS/test/*.tif'): 
    im = cv2.imread(filename, 0)
    im = testPreProcessing(im)
    # describe
    hist = extractLBP(im)
    # classify
    prediction = model.predict(hist.reshape(1,-1))
    print("Prediction: ",prediction)
    
```
