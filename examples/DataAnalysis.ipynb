# Data analysis 

```python
import pandas as pd
```

```python
# For this file we used Dice threshold 0.2, and the min area of a region of interest is 625
#data = pd.read_csv("feature-tables/train-data_1526149385.csv")

# For this file we used Dice threshold 0.2, and the min area of a region of interest is 2500
#data = pd.read_csv("feature-tables/train-data_1526205815.csv")
data = pd.read_csv("feature-tables/train-data_1526424532.csv")

data.describe()
```

```python
# data division into masses and not masses
data_masses = data[data["class_id"] == 1]

data_fp = data[data["class_id"] == -1]

print(f"Found {len(data_masses)} masses")
print(f"Found {len(data_fp)} false positives")
print(f"{round(100*len(data_masses)/len(data), 1)}% of the whole data is masses.")
```
